```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python
filename = "../data/CE_valid_E1.h5"

with h5py.File(filename, "r") as f:
    print("Keys: %s" % f.keys())
    print(f['valid'])
    print(f['valid'].keys())
    print(f['valid']['alpha'])
    print(f['valid']['beta'])
    print(f['valid']['gamma'])
    print(f['valid']['pde_250-100'])
    
    alpha = f['valid']['alpha'][()]
    beta = f['valid']['beta'][()]
    gamma = f['valid']['gamma'][()]
    
    pde_data = f['valid']['pde_250-200'][()]
```

```python
print(pde_data.shape)
```

```python
plt.plot(pde_data[0, 240, :])
plt.show()
```

```python
# Turn off matplotlib plot in Notebook
plt.ioff()
# Pass the ffmpeg path
plt.rcParams['animation.ffmpeg_path'] = 'ffmpeg'
```

```python
fig, ax = plt.subplots()
line, = ax.plot([])

data_index = 4

ax.set_xlim(0, pde_data.shape[2]-1)
ax.set_ylim(np.min(pde_data[data_index, :, :]) * 1.1, np.max(pde_data[data_index, :, :]) * 1.1)

x = np.arange(pde_data.shape[2])

def animate(frame_num):
    line.set_data((x, pde_data[data_index, frame_num, :]))
    return line

anim = FuncAnimation(fig, animate, frames=250, interval=10)

video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
anim.save('videos/burgers_' + str(data_index) + '_' + str(pde_data.shape[2]) + '.mp4')
plt.close()
```
